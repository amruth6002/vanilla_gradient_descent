```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
m_true,c_true=2,1
x=np.linspace(0,10,50)
y=m_true*x+c_true+np.random.normal(0,2,x.shape)
plt.scatter(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot of X vs Y")
plt.show()
```

```python
iterations=100
m,c=0.0,0.0
learning_rate=0.01
loss_his=[]
```

```python
for i in range(iterations):
    y_pred=m*x+c
    m_grad=(-2/len(x))*np.sum(x*(y-y_pred))
    c_grad = (-2/len(x))*np.sum((y-y_pred))

    m=m-learning_rate*m_grad
    c=c-learning_rate*c_grad

    loss=np.mean((y-y_pred)**2)
    loss_his.append(loss)
    if(i%10==0):
         print(f"Iteration {i}:Loss={loss:.4f},m={m:.4f},c={c:.4f}")
```

```python
plt.scatter(x, y,label="data points")
plt.plot(x,m*x+c,color="red",label=f"fitted line: y={m:.2f}x+{c:.2f}")
plt.legend()
plt.show()
```
